```python
import matplotlib.pyplot as plt
```

```python
INITIAL_GROWTH_RATE = .02
HALVING_TIME = 13
INITIAL_GDP = 1

def non_diminishing_gdp(year):
    return (1 + INITIAL_GROWTH_RATE) ** year

def diminishing_gdp(year):
    if year == 0:
        return INITIAL_GDP
    else:
        return (1 + growth_rate(year)) * diminishing_gdp(year - 1)

def growth_rate(year):
    return INITIAL_GROWTH_RATE * 0.5 ** (year / HALVING_TIME)
```

```python
def plot(title, y):
    plt.title(title)
    plt.plot(y)
    plt.ylabel("GDP")
    plt.xlabel("Year")
    plt.savefig(title + ".png")
    plt.show()
```

```python
plot("Non diminishing growth 100 yrs", [non_diminishing_gdp(i) for i in range(100)])
plot("Non diminishing growth 250 yrs", [non_diminishing_gdp(i) for i in range(250)])
plot("Non diminishing growth 1000 yrs", [non_diminishing_gdp(i) for i in range(1000)])
```

```python
plot("Diminishing growth 100 yrs", [diminishing_gdp(i) for i in range(100)])
plot("Diminishing growth 250 yrs", [diminishing_gdp(i) for i in range(250)])
plot("Diminishing growth 1000 yrs", [diminishing_gdp(i) for i in range(1000)])
```
